Term deposito merupakan bentuk investasi yang memiliki jangka waktu dan *return* tertentu dalam penyimpanannya. *Return* merupakan bunga dari hasil lama dan besaran penyimpanan yang dilakukan, bunga ini yang disebut sebagai penghasilan bagi mereka yang memberikan modal/dana. Sebagai pihak yang berperan penting atas jalannya roda perekenomian, sistem savings/simpanan merupakan salah satu aspek penting yang berpengaruh dalam pergerakan ekonomi dan finansial. Simpanan digunakan bank sebagai sumber dana/modal yang akan disalurkan kepada pihak/nasabah lain, dengan hal ini dapat membantu nasabah lain dalam mendorong perekonomian mereka yang juga bisa berdampak pada perekonomian negara dan jalan usahanya perusahaan (bank).

Deposito merupakan sistem simpanan dimana client/nasabah menitipkan asset cash mereka ke sebuah lembaga terpercaya atau disebut bank. Modal yang diberikan menjadi sumber perusahaan dalam menjalankan operasional sehari-hari dan sumber berkembang/bertumbuhnya perusahaan, sumber ini didapat dari keuntungan *return* berbentuk bunga oleh peminjam, dikenal dengan istilah *profit*. <br>
Salah satu cara untuk mendapatkan profit, maka perusahaan harus menarik banyak nasabah kepadanya, agar nasabah mendekati perusahaan, perusahaan perlu memiliki predikat serta record baik dan dapat dipercaya. Salah satunya dalam aspek *liquidity* dan *stability*. Bagaimana bank akan menjalankan operasional dan mendapatkan income dari peminjam jika tidak ada arus dana yang masuk. 

Oleh karena itu, perusahaan melakukan berbagai program berbasis tabungan salah satunya program deposito. `Melalui campaign, selain mempromosikan program dan menarik client/nasabah masuk, tentunya program harus menghasilkan keuntungan yang sepadan`.

### Metric Evaluation
Namun dalam penerapan *Machine Learning* memiliki kemungkinan *error* atas prediksi yang dirincikan seperti berikut:
<br>
<br>

|  | Tidak Menerima Tawaran (0) | Menerima Tawaran (1) |
| --- | --- | --- |
| **Tidak Menerima Tawaran (0)** | **True Negative (TN)** | **False Positif (FP)** |
|  | Model memprediksi *client* tidak menerima tawaran deposito, dan *client* tidak menerima tawaran. | Model memprediksi *client* menerima tawaran deposito, tetapi *client* tidak menerima tawaran deposito.|
| **Menerima Tawaran (1)** | **False Negatif (FN)** | **True Positif (TP)** |
|  | Model memprediksi *client* tidak menerima tawaran deposito, tetapi *client* menerima tawaran. | Model memprediksi *client* menerima tawaran deposito, dan *client* menerima tawaran deposito.|


---



| **Tipe error** | **Konsekuensi** |
| --- | --- |
| False Positive | Kehilangan biaya *marketing* dan pemborosan waktu. |
| False Negative | Kehilangan *potensial revenue* |

<br>
<br>

  - Apabila saat itu (2014) rata rate deposito: 0.2%[<sup>2</sup>](https://tradingeconomics.com/portugal/interest-rate), rata rate pinjaman: 5.52%[<sup>3</sup>](https://tradingeconomics.com/portugal/bank-lending-rate) dalam 1 tahun.

````python
def detect_outlier(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3-q1
    lower = q1 - (1.5*iqr)
    upper = q3 + (1.5*iqr)

    return df[(df[column] < lower) | (df[column] > upper)]

for i in df.select_dtypes('number'):
    data = detect_outlier(df, i)
    if len(data) == 0:
        print(f'{i} has no outlier detected')
    else:
        total = len(data)*100/len(df)
        positive = len(data[data['y'] == 1])*100/len(df[df['y'] == 1])
        negative = len(data[data['y'] == 0])*100/len(df[df['y'] == 0])
        print(f'{i} has {total.__round__(2)}% outlier detected with target 0 contains {negative.__round__(2)}% and target 1 contains {positive.__round__(2)}%')
````

Data campaign dan duration memiliki beberapa gap outlier yang cukup jauh dari data pusat, namun outlier ini akan dibiarkan karena tidak ada yang mencurigakan dari data outlier tersebut. <br>

1. **Young Adults** (17-25) : Menyasar individu muda yang baru memasuki dunia dewasa, biasanya berfokus pada pendidikan dan awal karier.
2. **Emerging Adults** (26-35) : Fase peralihan, mendalami karier dan membangun relasi yang lebih kuat.
3. **Mid Adults** (36-45) : Segmen ini umumnya berada di puncak karier dan menghadapi tanggung jawab keluarga yang lebih besar.
4. **Established Adults** (46-55) : Individu yang berada pada fase stabil dalam karier dan kehidupan pribadi, sering kali fokus pada persiapan pensiun.
5. **Late Adults** (56-65) : Individu yang mendekati pensiun, cenderung mulai memikirkan rencana pensiun dan perjalanan hidup.
6. **Elderly** (over 65) : Kelompok ini biasanya sudah pensiun dan lebih fokus pada aktivitas rekreasi dan kesehatan.

[Reference](https://www.researchgate.net/publication/338842581_A_cross-sectional_study_on_interference_control_age_affects_reactive_control_but_not_proactive_control)

````python
# https://www.flexjobs.com/blog/post/5-stages-of-a-typical-career/?msockid=05055a6b55bb66ed18d34e4e54ed675b
def age_range(age):
  if age <= 25:
    return 'Young Adults' 
  elif age <= 35:
    return 'Emerging Adults'
  elif age <= 45:
    return 'Mid Adults'
  elif age <= 55:
    return 'Established Adults'
  elif age <= 65:
    return 'Late Adults'
  else:
    return 'Elderly'

df['age_range'] = df['age'].apply(lambda x: age_range(x))
````

#### Default

````python
contingency_table_default = (pd.crosstab(df['default'], df['y'], normalize='index')*100).round(3)
````

````python
plt.figure(figsize=(12,4))

ax = sns.countplot(data=df, x='default', hue='y', order=contingency_table_default.index)
for p in ax.patches:
    ax.annotate(f'{round(p.get_height()*100/len(df),2)}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', fontsize=9, color='black', xytext=(0, 5),
                    textcoords='offset points')
ax.tick_params(axis='x', rotation=45)
for spine in ax.spines.values():
    spine.set_visible(False)

ax2 = ax.twinx()
ax2.plot(contingency_table_default.index, contingency_table_default[1], color='red', marker='o', label='Yes Percentage', linestyle='--')
ax2.set_ylabel("Yes Percentage (%)", color='red')
ax2.tick_params(axis='y', labelcolor='red') # or set_xtickslabel 

plt.setp(ax2.spines.values(), visible=False)
plt.title("Acceptance Rate of Client by Default")
plt.tight_layout();
````

Berdasarkan grafik diatas, dapat disimpulkan bahwa:

1.  Status default "no" memiliki persentase keberhasilan campaign yang tinggi, yaitu sekitar 11%

    Alasan meningkatnya tingkat keberhasilan campaign kepada mereka yang belum pernah mengalami default (kemacetan pembayaran):
       - **Financial Stability**: Individu yang tidak pernah memiliki riwayat kegagalan pembayaran pinjaman dapat dikatakan situasi financial individu tersebut memiliki stabilitas yang baik. Hal ini memungkinkah mereka untuk menerima penawaran dikarenakan kemungkinanan tidak adanya kecenderungan kekhawatiran terhadap kondisi finansial nasabah, sehingga bukanlah masalah bagi nasabah untuk menaruh sebagian aset mereka kedalam deposito.

2.  Status default "yes" memiliki persentase keberhasilan campaign yang rendah, yaitu < 1%

    Alasan mungkin untuk menurunnya tingkat keberhasilan campaign pada "yes":
       - **Financial Stability**: Individu dengan riwayat kegagalan pembayaran kurang lebihnya menggambarkan kondisi finansial nasabah yang cenderung tidak stabil, maka tidak heran jika nasabah lebih memprioritaskan kas mereka kepada pemenuhan kebutuhannya terlebih dahulu daripada melakukan penyimpanan ke dalam bank (karena nasabah membutuhkan akses dana segera).
       
Individu dengan status default "yes", cenderung akan memiliki score credit yang buruk apabila individu mengalami kemacetan pembayaran dalam jangka waktu yang panjang dan terjadi dalam tempo yang sering, maka perusahaan dapat melakukan pendekatan yang bisa `memulihkan atau meningkatkan score credit mereka kembali apabila nasabah melakukan pembukaan deposito`.<br>

Misalkan nasabah yang sedang berada dalam kemacetan pembayaran (asumsi kartu kredit), hanya perlu membayar biaya bunga (asumsi 2%) keterlambatan dan sebagian % biaya pokok belanja sedangkan sisa biaya pokok lainnya dicicil dan dialokasikan kedalam deposito dengan minimum jangka n-month dan return bunga yang didapatkan x-rate% (jangka waktu dan return bunga disesuaikan dengan kalkulasi keuntungan perusahaan), `perusahaan dapat melakukan pemotongan deposito client sekian persen sebagai pemenuhan biaya pokok belanja yang sudah digunakan oleh client dan renewal deposito` apabila keuntungan belum tercapai dan deposito diberikan `fleksibilitas untuk melakukan penambahan kas tiap bulannya`.

Keuntungan bagi *defaulted-client*, client tidak merasa terbebani dengan biaya yang dikeluarkan untuk pelunasan kredit, namun client akan merasa biaya pokok belanja yang biasanya dikeluarkan cuma-cuma untuk pelunasan kredit dapat mejadi sumber income tambahan (dari return bunga) dikarenakan biaya pokok tersebut dialokasikan sebagai tabungan deposito.<br>
Bagi perusahaan, dengan menarik client yang memiliki catatan credit buruk untuk mengalokasikan kasnya kedalam deposito dan pemberian bunga kepada client dapat meminimalisir potensi kerugian apabila client memutuskan pembayaran**. 

*Dengan adanya penerapan renewal serta pemotongan % deposito dan disesuaikan dengan kalkulasi jangka waktu dan return bunga yang tepat, maka perusahaan seharusnya tidak akan mengalami kerugian namun mengalami peningkatan revenue dari client yang melakukan pembayaran secara berkala (misalnya setiap n-bulan) kedalam deposito*. 

**Nasabah yang disarankan adalah nasabah yang memiliki record credit score 'low-medium' (atau status buruk hingga sangat buruk), dikarenakan nasabah memiliki potensi untuk memutus pembayaran, strategi pendekatan atas dapat menjadi pencegah atau meminimalisir kemungkinan hal tersebut terjadi<br>
